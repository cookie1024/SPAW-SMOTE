```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath = "E:\PyCharmProject\SPWSYN3.1\evaluation\SVM.xlsx"
data_frame=pd.read_excel(filepath, usecols="B:J")
#新建一个工作簿
data = []
#print(data_frame)
for indexs in data_frame.index:
    data.append(indexs)
    data[indexs]=[]
    t = len(data_frame.loc[indexs].values[:])
    for i in range(t):
        data[indexs].append(data_frame.loc[indexs].values[i])
data = np.array(data)
print(data)
```

```python
#print(data.shape[0])
num1 = np.zeros((3,8), dtype=np.int)
num2 = np.zeros((3,8), dtype=np.int)
num3 = np.zeros((3,8), dtype=np.int)
#print(num1)
for i in range(data.shape[0]):
    k = i%3
    #print(k)
    for j in range(data.shape[1]-1):
        if data[i][j] > data[i][8]:
            num1[k][j] = num1[k][j] + 1
        elif data[i][j] < data[i][8]:
            num2[k][j] = num2[k][j] + 1
        else:
            num3[k][j] = num3[k][j] + 1
print(num1)
print(num2)
print(num3)
```
